```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

print('Gamma Correction pada Citra')
print('-------------------------')

try:
    gamma = int(input('Masukkan nilai Gamma: '))
except ValueError:
    print('Error, bukan angka')

original = cv.imread('/content/drive/MyDrive/PCVK/Images/female.tiff')
gamma_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      gamma_image[y,x,c] = np.clip(255 * pow  (original[y,x,c] / 255, (1 / gamma)), 0, 255)

final_frame = cv.hconcat((original, gamma_image))
cv2_imshow(final_frame)
```

```python
#membaca gambar dalam format tiff
original = cv.imread('/content/drive/MyDrive/PCVK/Images/female.tiff', cv.IMREAD_GRAYSCALE)
#membuat array numpy dengan ukuran yang sama untuk menyimpan citrs kedalaman
depth_image = np.zeros(original.shape, original.dtype)
#mengatur jumlah bit kedalaman yang akan digunkana dalam simulasi
bith_depth=2
#menghitung nilai level atau rentang intensitas yang akan digunakan dalam pembuatan kedalaman
level = 255 / (pow(2,bith_depth)-1)

#memulai loop
for y in range(original.shape[0]):
  #setiap piksel dalam baris gambar aslidibagi dengan nilai level,kemudian dibulatkan dengan np.round dan dikalikan lagi dengan nilai level
  depth_image[y] = np.clip(np.round(original[y]/level)*level, 0, 255)
#menggabungkan gambar asli dengan citra kedalaman
final_frame = cv.hconcat((original, depth_image))
#menambahkan frame
final_frame = cv.cvtColor(final_frame, cv.COLOR_BGR2RGB)
plt.imshow(final_frame)
```

```python
# 1. Jumlah Citra di Average - 5

import glob
from math import log10, sqrt

original = cv.imread('/content/drive/MyDrive/PCVK/Images/galaxy.jpg')
cv_img = []
for img in glob.glob('/content/drive/MyDrive/PCVK/Images/noises/*.jpg'):
    n = cv.imread(img)
    cv_img.append(n)

dst = cv_img[0]
noise_number = 4
for i in range(noise_number):
    if i == 0:
        pass
    else:
        alpha = 1.0 / (i + 1)
        beta = 1.0 - alpha
        dst = cv.addWeighted(cv_img[i], alpha, dst, beta, 0.0)

def PSNR(original, compressed):
    mse = np.mean((original - compressed) ** 2)
    if(mse == 0):
      return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse))
    return psnr

compressed = dst
psnr = PSNR(original, compressed)

print('Nilai PSNR adalah',psnr,'(dB)')
galaxy_frame = cv.hconcat((original, dst))
cv2_imshow(galaxy_frame)
```

```python
#Operator AND
img_t = cv.imread('/content/drive/MyDrive/PCVK/Images/couple.tiff')

mask = np.zeros(img_t.shape, dtype=np.uint8)
mask = cv.circle(mask, (70, 40), 40, (255,255,255), -1)
mask2 = cv.circle(mask, (195, 40), 40, (255,255,255), -1)

result = cv.bitwise_and(img_t,mask)

result[mask==0] = 255
final_frame = cv.hconcat((img_t, mask, result))
cv2_imshow(final_frame)
```

```python
#Operator NAND
img_t = cv.imread('/content/drive/MyDrive/PCVK/Images/couple.tiff')

mask = np.zeros(img_t.shape, dtype=np.uint8)
mask = cv.circle(mask, (70, 40), 40, (255,255,255), -1)
mask2 = cv.circle(mask, (195, 40), 40, (255,255,255), -1)

# Mengganti operasi bitwise OR menjadi NAND
result = cv.bitwise_not(cv.bitwise_and(cv.bitwise_not(img_t), cv.bitwise_not(mask)))

# Mengubah nilai piksel yang tidak termasuk dalam mask menjadi 0
result[mask==0] = 0

final_frame = cv.hconcat((img_t, mask, result))
cv2_imshow(final_frame)
```
